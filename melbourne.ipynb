```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("Melbourne_housing_FULL.csv")
df
```

```python
df.nunique()
```

```python
print(df.shape)
```

```python

df= df[['Suburb','Rooms','Type','Method','SellerG','Distance','Bedroom2','Bathroom','Car','Landsize','BuildingArea','CouncilArea','Regionname','Propertycount','Price']]
df
```

```python
df= df.dropna()
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df1= pd.get_dummies(df,dtype="int")
df1
```

```python
X= df1.drop(columns=['Price'])
Y= df1[['Price']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Lasso
model1= Lasso(alpha=25,max_iter=100,tol=0.1)
model1.fit(X_train,Y_train)
```

```python
model1.score(X_train,Y_train)
```

```python
model1.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Ridge
model2= Ridge(alpha=50,max_iter=100,tol=0.1)
model2.fit(X_train,Y_train)
```

```python
model2.score(X_train,Y_train)
```

```python
model2.score(X_test,Y_test)
```
